We have dealt with the missing values on Adult mortality row

```python
new_df = df_drop
```

```python
new_df.head()
```

#### Dealing with the Alcohol column, according to world's data, there are countries that do not consume alcohol, to deal with this null values, we will impute these values with the mean by country; this will involve:
- Grouping by country
- Calculating the mean of alcohol for each country
- fill in the values

  #### or we have an option of using K - nearest neighbor but that'll be too easy.
  I am not filling in the value by getting the mean of the whole column and filling it in because it raises the risk of getting outliers. 

```python
mean_alcohol_by_country = new_df.groupby('Country')['Alcohol'].transform('mean')
new_df['Alcohol'] = new_df['Alcohol'].fillna(mean_alcohol_by_country)
new_df
```

```python
new_df.isna().sum()
```

```python
#Alcohol still has 16 null values, let us view them
alcohol = new_df[new_df['Alcohol'].isnull()]
alcohol
```

```python
#since Alcohol has a consistent NA value in south sudan, we would fill it with the 0.001 value
new_df ['Alcohol'] = new_df['Alcohol'].fillna(0.001)
new_df
```

```python
new_df.isna().sum()
```

```python
#now we follow same pattern for hepatitis and BMI column
hepatitis_alcohol_by_country = new_df.groupby('Country')['Hepatitis B'].transform('mean')
new_df['Hepatitis B'] = new_df['Hepatitis B'].fillna(hepatitis_alcohol_by_country)
new_df
```

```python
new_df.isna().sum()
```

Our missing values for hepatitis B has reduced to 144

```python
hepatitis = new_df[new_df['Hepatitis B'].isnull()]
hepatitis
```

Considering the countries with no case of Hepatitis B are developed countries, the values will be filled with 0

```python
new_df ['Hepatitis B'] = new_df['Hepatitis B'].fillna(0)
new_df
```

```python
new_df.columns
```

```python
bmi_alcohol_by_country = new_df.groupby('Country')[' BMI '].transform('mean')
new_df[' BMI '] = new_df[' BMI '].fillna(bmi_alcohol_by_country)
new_df.isnull().sum()
```

```python
null_bmi= new_df[new_df[' BMI '].isnull()]
null_bmi
```

```python
for i in [" BMI ","Income composition of resources", "Polio", "GDP"]:
    new_df[i].fillna(new_df[i].median(), inplace=True)

#fill in missing valiues with median of the row.
```

```python
new_df.isnull().sum()
```

The rest of the missing values in the dataset will be filled using K-Nearest neighbor

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer()
```

```python
for i in new_df.select_dtypes(include="number").columns:
    new_df[i] = imputer.fit_transform(new_df[[i]])
```
